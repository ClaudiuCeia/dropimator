```python
import pandas as pd
import numpy as np
from ast import literal_eval

from sklearn.metrics import classification_report

datafile_path = 'product_embeddings.csv'

df = pd.read_csv(datafile_path)
df["embedding"] = df['embedding'].apply(literal_eval).apply(np.array)

df
```

```python
from openai.embeddings_utils import cosine_similarity, get_embedding
from sklearn.metrics import PrecisionRecallDisplay

def evaluate_embeddings_approach(product_embeddings, labels, embedding_model):
    label_embeddings = [get_embedding(label, engine=embedding_model) for label in labels]

    def label_score(product_embeddings, label_embeddings):
        assigned_labels = []
        for product in product_embeddings:
            scores = [cosine_similarity(product, label) for label in label_embeddings]
            assigned_labels.append(labels[scores.index(max(scores))])
        return assigned_labels

    assigned_labels = label_score(product_embeddings, label_embeddings)
    return assigned_labels
    
product_embeddings = df['embedding']    
labels = [
    'Proteine',
    'Aminoacizi',
    'Vitamine si Minerale',
    'Batoane si Gustari Fitness',
    'Suplimente pentru slabit',
    'Performanta/Stimulatoare',
    'Pre-Workout',
    'Creatina',
    'Imbracaminte si acesorii pentru sala',
    'Masa musculara'
]
embedding_model = "text-embedding-ada-002"

print(evaluate_embeddings_approach(product_embeddings, labels, embedding_model));
```
